```python
import pandas as pd
!pip install geopy
from geopy.geocoders import Nominatim 
```

```python
df = pd.read_csv('/content/drive/MyDrive/PROJECT_senior it support center/데이터/주민센터 주소.csv')
df.head()
```

```python
geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)

def geocoding(address):
    try :
        geo = geolocoder.geocode(address)
        crd = (str(geo.latitude), str(geo.longitude))
        return crd
    except :
        return (0, 0)
```

```python
lats = []
logs = []

for address in df['주소'] :
    crd = geocoding(address)

    lats.append(crd[0])
    logs.append(crd[1])

lats = pd.Series(lats)
logs = pd.Series(logs)
```

```python
lats
```

```python
df = pd.concat([df, lats], axis=1)
df = pd.concat([df, logs], axis=1)
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
df.columns=['시설명', '주소', '위도', '경도']
```

```python
import pandas as pd
import numpy as np
import pyproj
from pyproj import Transformer
```

```python
df.head()
```

```python
# 서울역 위도, 경도((37.553230, 126.972551) (예시)
transformer = Transformer.from_crs("epsg:4326", "epsg:5178")
transformer.transform(37.553230, 126.972551)
```

```python
# transformer.transform(37.560927, 127.193877)
list_utmk = []
for i in range(len(df)):
    list_utmk.append(transformer.transform(df['위도'][i], df['경도'][i]))

list_utmk
```

```python
df_utmk = pd.DataFrame(list_utmk)
df_utmk.columns = ['위도utmk', '경도utmk']
df_utmk
```

```python
main = pd.concat([df, df_utmk], axis=1)
main
```
