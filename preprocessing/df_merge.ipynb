# 주변특성, 대중교통

```python
main = pd.read_excel('/content/drive/MyDrive/PROJECT_senior it support center/데이터/main_df.xlsx',
                index_col=0)
main.head()
```

```python
# 주변특성 파생변수 생성
# 시니어들이 자주 이용하는 시설들의 개수를 합해서, 주변에 시니어를 위한 시설이 얼마나 있는지 확인
main['주변특성'] = main['도서관(500)'] + main['배움터(500)'] + main['병원(500)'] + main['공원(500)']


# 대중교통 파생변수 생성 - 주변에 교통이 얼마나 편리한지 보기 위한 변수.
main['대중교통'] = main['버스(100)'] + main['지하철(250)']
main.head()
```

```python
final = main[['시설명', '주변특성', '대중교통']]
final.head()
```

```python
final['동'] = final['시설명'].map(lambda x : x.split()[0])
final.head()
```

```python
final = final[['동', '시설명', '주변특성', '대중교통']] # 칼럼 순서정리
final.head()
```

```python
final[final['동'].str.contains('-')]  # merge 할때 표기방식의 차이로 문제가 발생할 행들.
```

```python
final = final.replace('금호2-3가동', '금호2.3가동')
```

```python
final.info()
```

# 디지털정보지수

```python
media = pd.read_csv('/content/drive/MyDrive/PROJECT_senior it support center/데이터/media_df.csv',
                    index_col=0)
media.head()
```

```python
media = media.replace('상일1동', '상일동')
media [media ['동'].str.contains('상일')]
```

```python
final_df = pd.merge(final, media, how='left', on=['동'])
final_df
```

```python
final_df[final_df['동']=='신사동']
```

```python
final_df.drop_duplicates(inplace=True) #중복행 제거
final_df[final_df['동']=='신사동']
```

```python
final_df.info()  # null값이 없이, merge가 잘 됐다.
```

```python
final_df = final_df[['구', '동', '시설명', '주변특성', '대중교통', '디지털정보지수']]
final_df
```
