# 인구특성 변수 전처리

- 동별 60세 이상 시니어 인구 수 구하기
- 동별 시니어 비율 구하기

## Imports

```python
# 필요 라이브러리 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글폰트 설정
!pip install koreanize-matplotlib
import koreanize_matplotlib

# 폰트 선명하게
%config InlineBackend.figure_format = 'retina'
```

```python
people_raw = pd.read_excel('/content/drive/MyDrive/PROJECT_senior it support center/데이터/인구특성.xlsx')
people_raw.head()

# 엑셀로 동별 전체인구, 동별 시니어 인구(60세 이상 인구), 동별 시니어 비율(시니어 인구 / 전체인구)를 구함.
```

```python
# 사용할 컬럼만 뽑기

# 인덱스 컬럼 : 구, 동
people1 = people_raw.iloc[:, :2]
people1.head()
```

```python
# 동별 전체인구, 동별 시니어 수, 동별 시니어 비율 컬럼만 뽑기
people2 = people_raw.iloc[:, -3:]
people2.head()
```

```python
# 사용할 인구 데이터셋 완성 : people1 + people2
people = pd.concat([people1, people2], axis=1)
people.head()
```

```python
# 데이터 저장
people.to_csv('/content/drive/MyDrive/PROJECT_senior it support center/데이터/people.csv')
```
