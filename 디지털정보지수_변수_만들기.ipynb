- 행정동별 정보화격차를 고려하기 위해 만든 변수.
- 디지털정보지수 변수 구한 방법 :
    - 각 행정동 별 (50대 인구 * 50대 디지털 정보화지수) + (60대 인구 * 60대 정보화지수) ... 이런 식으로 구한 변수.
- 하지만 잘못 사용한 변수임.
    - 1) 디지털정보화지수가 높으면, 그 지역에 it 시설을 설치하겠다는 건지? 지금의 흐름대로라면 그렇게 된다.
    - 2) 하지만, 프로젝트의 취지대로 라면, 오히려 디지털정보화지수가 낮은 지역에 it 지원시설을 설치해야 함.
    - 3) 또한, 50대 이상 부터 정보화지수에 반영한 것도 근거가 없음.
- 위의 이유들로, 프로젝트를 다시 해보게 된다면, 이 변수 반영여부는 재고려할 필요가 있음.

# Imports

```python
import pandas as pd
```

```python
# 데이터 로드
media = pd.read_excel('/content/drive/MyDrive/PROJECT_senior it support center/데이터/전처리 전 데이터/인구특성.xlsx')
media.head()
```

```python
# 50대, 60대, 70대 이상 인구수 * 각 연령별 디지털정보지수를 할 예정. 

# 일단 50세 이상만 슬라이싱, 불필요한 컬럼 drop
media_0 = media.iloc[:, :2]

media = media.loc[:, '50~54세':]
media = media.iloc[:, :-3]
media.head()
```

```python
# 50대, 60대, 70대 이상 인구수 만들기
media['50대'] = media.iloc[:, :2].sum(axis=1, numeric_only=True)
media['60대'] = media.iloc[:, 2:4].sum(axis=1, numeric_only=True)
media['70대 이상'] = media.iloc[:, 4:11].sum(axis=1, numeric_only=True)
media.head()
```

```python
# 사용할 마지막 3 컬럼만 슬라이싱
media = media.iloc[:, -3:]
media.head()
```

```python
# 50대, 60대, 70대 이상 디지털 역량수준
dgt_level = [0.92, 0.593, 0.224]

# 디지털 역량수준 * 각 연령대별 인구 수
result = round(media * dgt_level, 2)
result['디지털정보지수'] = result.sum(axis=1)

# 최종 df 만들기
media_df = pd.concat([media_0, result['디지털정보지수']], axis=1)
media_df
```

```python
media_df.to_csv('/content/drive/MyDrive/PROJECT_senior it support center/데이터/media_df.csv')
```
