# 도서관

```python
library = pd.read_excel('/content/drive/MyDrive/PROJECT_senior it support center/데이터/전처리 전 데이터/공공도서관 위도경도 변환.xlsx',
                        index_col=0)
library.head()
```

```python
# #  위도, 경도만 슬라이싱
# lib_utmk = library.loc[:, ['위도UTMK', '경도UTMK']]
# lib_utmk.head()
```

```python
main = pd.read_csv('/content/drive/MyDrive/PROJECT_senior it support center/데이터/main_utmk.csv',
                   index_col=0)
main['도서관(500)'] = 0 # 도서관(500) 개수 들어갈 칼럼 생성
main.head()
```

```python
# main 데이터프레임의 위경도
x = main['위도utmk']
y = main['경도utmk']

# 도서관 데이터프레임의 위경도
lib_x = library['위도UTMK']
lib_y = library['경도UTMK']
```

```python
# 각 주민센터로부터 500이내 거리에 있는 도서관 개수 구하기

# 각 주민센터로부터 도서관 거리 구해서, 500이내면 개수 카운트
for i in range(len(main)):
    for j in range(len(library)):
        value = False
        value = np.sqrt((x[i] - lib_x[j])**2 + (y[i] - lib_y[j])**2) <= 500
        if value == True:
            main['도서관(500)'][i] += int(value)
        else:
            pass
main
```
